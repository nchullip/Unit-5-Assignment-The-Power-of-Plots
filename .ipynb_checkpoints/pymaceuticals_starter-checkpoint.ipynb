```python
# Dependencies and Setup
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# Hide warning messages in notebook
import warnings
warnings.filterwarnings('ignore')

# File to Load
mouse_drug_data_to_load = "data/mouse_drug_data.csv"
clinical_trial_data_to_load = "data/clinicaltrial_data.csv"

# Read the Mouse and Drug Data and the Clinical Trial Data
mouse_drug_df=pd.read_csv(mouse_drug_data_to_load)
clinical_trial_df=pd.read_csv(clinical_trial_data_to_load)

# Combine the data into a single dataset
merged_df=pd.merge(clinical_trial_df,mouse_drug_df, on="Mouse ID", how="outer")

# Display the data table for preview
merged_df.head()
```

## Tumor Response to Treatment

```python
# Store the Mean Tumor Volume Data Grouped by Drug and Timepoint
merged_grouped_df = merged_df.groupby(["Drug", "Timepoint"])

tumor_mean = merged_grouped_df["Tumor Volume (mm3)"].mean()

# Convert to DataFrame
tumor_response_df = pd.DataFrame(tumor_mean)

# Preview DataFrame
tumor_response_summary_df = tumor_response_df.copy()
tumor_response_summary_df.reset_index()
```

```python
# Store the Standard Error of Tumor Volumes Grouped by Drug and Timepoint
standard_error=merged_grouped_df["Tumor Volume (mm3)"].sem()

# Convert to DataFrame
standard_error_df=pd.DataFrame(standard_error)

# Preview DataFrame
standard_error_summary_df=standard_error_df.copy()
standard_error_summary_df.reset_index().head()
```

```python
# Minor Data Munging to Re-Format the Data Frames
summary_tumor_volume_df=pd.pivot_table(tumor_response_summary_df, index="Timepoint", columns="Drug", values="Tumor Volume (mm3)")

# Preview that Reformatting worked
summary_tumor_volume_df
```

```python
# Generate the Plot (with Error Bars)
plt.errorbar(summary_tumor_volume_df.index, summary_tumor_volume_df["Capomulin"], xerr=0.75, yerr=0.75,
             color='r', marker='o', markersize=4, linestyle='--', linewidth=1)

plt.errorbar(summary_tumor_volume_df.index, summary_tumor_volume_df["Infubinol"], xerr=0.75, yerr=0.75,
             color='b', marker='^', markersize=4, linestyle='--', linewidth=1)

plt.errorbar(summary_tumor_volume_df.index, summary_tumor_volume_df["Ketapril"], xerr=0.75, yerr=0.75,
             color='g', marker='s', markersize=4, linestyle='--', linewidth=1)

plt.errorbar(summary_tumor_volume_df.index, summary_tumor_volume_df["Placebo"], xerr=0.75, yerr=0.75,
             color='black', marker='D', markersize=4, linestyle='--', linewidth=1)

x_lim = len(summary_tumor_volume_df.index)
plt.title("Tumor Response to Treatment")
plt.xlabel("Time (Days)")
plt.ylabel("Tumor Volume (mm3)")
plt.legend(loc='upper left')
plt.grid()

# Save the Figure
plt.savefig("Output/tumor_response.png")
```

```python
# Show the Figure
plt.show()
```

## Metastatic Response to Treatment

```python
# Store the Mean Metastatic Response Data Grouped by Drug and Timepoint
merged_grouped_df=merged_df.groupby(["Drug", "Timepoint"])

metastatic_response_mean=merged_grouped_df["Metastatic Sites"].mean()

# Convert to DataFrame
metastatic_response_df=pd.DataFrame(metastatic_response_mean)

# Preview DataFrame
metastatic_response_summary_df=metastatic_response_df.copy()
metastatic_response_summary_df.reset_index()
```

```python
# Store the Standard Error associated with Met. Sites Grouped by Drug and Timepoint 
standard_error_ms=merged_grouped_df["Metastatic Sites"].sem()

# Convert to DataFrame
standard_error_ms_df=pd.DataFrame(standard_error_ms)

# Preview DataFrame
standard_error_ms_summary_df=standard_error_ms_df.copy()
standard_error_ms_summary_df.reset_index()
```

```python
# Minor Data Munging to Re-Format the Data Frames
summary_metastatic_response_df=pd.pivot_table(metastatic_response_summary_df, index="Timepoint", columns="Drug", values="Metastatic Sites")

# Preview that Reformatting worked
summary_metastatic_response_df
```

```python
# Generate the Plot (with Error Bars)
plt.errorbar(summary_metastatic_response_df.index, summary_metastatic_response_df["Capomulin"], xerr=0.3, yerr=0.3,
             color='r', marker='o', markersize=4, linestyle='--', linewidth=1)

plt.errorbar(summary_metastatic_response_df.index, summary_metastatic_response_df["Infubinol"], xerr=0.3, yerr=0.3,
             color='b', marker='^', markersize=4, linestyle='--', linewidth=1)

plt.errorbar(summary_metastatic_response_df.index, summary_metastatic_response_df["Ketapril"], xerr=0.3, yerr=0.3,
             color='g', marker='s', markersize=4, linestyle='--', linewidth=1)

plt.errorbar(summary_metastatic_response_df.index, summary_metastatic_response_df["Placebo"], xerr=0.3, yerr=0.3,
             color='black', marker='D', markersize=4, linestyle='--', linewidth=1)

x_lim = len(metastatic_response_summary_df.index)
plt.title("Metastatic Spread During Treatment")
plt.xlabel("Treatment Duration (Days)")
plt.ylabel("Met. Sites")
plt.legend(loc='upper left')
plt.grid()

# Save the Figure
plt.savefig("Output\metastatic_response.png")
```

## Survival Rates

```python

# Store the Count of Mice Grouped by Drug and Timepoint (W can pass any metric)
merged_grouped_df=merged_df.groupby(["Drug", "Timepoint"])

mice_count=merged_grouped_df["Mouse ID"].count()

# Convert to DataFrame
mice_count_df=pd.DataFrame(mice_count)

# Preview DataFrame
mice_count_summary_df=mice_count_df.copy()

mice_count_summary_df.reset_index().head()
```

```python
# Minor Data Munging to Re-Format the Data Frames
summary_mice_count_df=pd.pivot_table(mice_count_df, index="Timepoint", columns="Drug", values="Mouse ID")

# Preview that Reformatting worked
summary_mice_count_df
```

```python
# Generate the Plot (Accounting for percentages)
survival_rate= (summary_mice_count_df/(summary_mice_count_df.iloc[0,:])*100)

plt.plot(survival_rate.index, survival_rate["Capomulin"],color='r', marker='o', markersize=4, linestyle='--', linewidth=1)

plt.plot(survival_rate.index, survival_rate["Infubinol"],color='b', marker='^', markersize=4, linestyle='--', linewidth=1)

plt.plot(survival_rate.index, survival_rate["Ketapril"],color='g', marker='s', markersize=4, linestyle='--', linewidth=1)

plt.plot(survival_rate.index, survival_rate["Placebo"],color='black', marker='D', markersize=4, linestyle='--', linewidth=1)


plt.xlim = len(metastatic_response_summary_df.index)
plt.title("Survival During Treatment")
plt.xlabel("Treatment Duration (Days)")
plt.ylabel("Survival Rate (%)")
plt.legend(loc='lower left')
plt.grid()

# Save the Figure
plt.savefig("Output/Survival During Treatment.png")
```

## Summary Bar Graph

```python
# Calculate the percent changes for each drug
percent_change=((summary_tumor_volume_df.iloc[9,:])/(summary_tumor_volume_df.iloc[0,:])*100)-100

# Display the data to confirm
percent_change
```

```python
# Store all Relevant Percent Changes into a Tuple
tuple= ("-19.475303", "42.516492", "46.123472", "57.028795", "53.923347", "51.29796", "47.241175", "22.3209", "52.085134"
        "46.579751")

drug_summary = ["Capomulin","Infubinol","Ketapril","Placebo"]
change_summary = [-19.475303, 46.123472, 57.028795, 51.29796]

plt.ylim(-30, 70)

# Orient widths. Add labels, tick marks, etc. 
colors= ["green" if (change < 0) else "red" for change in percent_change]

graph= plt.bar(drug_summary,change_summary, width=-1, align='edge',color=colors)

plt.title("Tumor Change Over 45 Day Treatment")
plt.ylabel("% Tumor Volume Change")
ax=plt.subplot()

# Use functions to label the percentages of changes
def autolabel(rects, ax):
# Get y-axis height to calculate label position from.
    (y_bottom, y_top) = ax.get_ylim()
    y_height = y_top - y_bottom

    for rect in rects:
        height = rect.get_height()

        # Fraction of axis height taken up by this rectangle
        p_height = (height / y_height)

        # If we can fit the label above the column, do that;
        # otherwise, put it inside the column.
        if p_height > 0:
            label_position = 0
        else:
            label_position = -5

        ax.text(rect.get_x() + rect.get_width()/2., label_position, '%d' % int(height) +"%", color="white", ha='center', va='bottom')
        
# Call functions to implement the function calls
autolabel(graph,ax)

# Show the Figure
plt.grid()

# Save the Figure
plt.savefig("Output/Summary Bar Graph.png")
```
